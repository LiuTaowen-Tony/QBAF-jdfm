```python
import numpy as np
import pandas as pd
from utils import fuzzy_binning
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 1000)
data.sort()
data = data.reshape(-1, 1)

data_pd = pd.DataFrame(data)

plt.scatter(data, np.zeros_like(data))


binned, _= fuzzy_binning(data_pd, n_bins=4, feature_names=["test"])
binned = binned.to_numpy()

class1 = binned[:, 0]
class2 = binned[:, 1]
class3 = binned[:, 2]
class4 = binned[:, 3]

plt.plot(data, class1)
plt.plot(data, class2)
plt.plot(data, class3)
plt.plot(data, class4)
# plt.plot(class1, data)
# plt.plot(class2, data)
# plt.plot(class3, data)

```
